```python
# Customer Demographics Visualization

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("D:\\mini project\\boutique.csv")

# Age distribution
plt.hist(df['age'], bins=10)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

```

```python
top_cities = df['City'].value_counts().head(10)  # Top 10 cities
top_cities.plot(kind='barh')  # Horizontal bar plot to show cities on the y-axis
plt.title('Top 10 Cities by Number of Customers')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

```python
# sales performance by category or product

# Top 10 selling products by total amount
top_products = df.groupby('product name')['Total_Amount'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar')
plt.title('Top 10 Selling Products by Total Revenue')
plt.ylabel('Total Revenue')
plt.show()

import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing sales data with 'product name' and 'Total_Amount'

# Grouping by product name and summing the total amount for each product
top_products = df.groupby('product name')['Total_Amount'].sum().sort_values(ascending=False).head(10)

# List of colors (you can customize or use a color map)
colors = plt.cm.Paired(range(10))  # 'Paired' is a colormap that provides a set of distinct colors

# Plotting the top 10 selling products with custom colors
top_products.plot(kind='bar', color=colors, figsize=(10,6))
plt.title('Top 10 Selling Products by Total Revenue', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Total Revenue', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate product names for better readability
plt.tight_layout()  # Adjust layout so labels are not cut off
plt.show()



```

```python
# Category-wise sales distribution
category_sales = df.groupby('category')['Total_Amount'].sum()
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Category')
plt.ylabel('')
plt.show()
```

```python
# Monthly or Yearly Sales Trends

# Convert the 'Date' column to datetime format, inferring the format
df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format=True, errors='coerce')

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Aggregate by month for monthly trend analysis
monthly_data = df.resample('M').sum()

# Monthly sales trend
monthly_sales = df.resample('M')['Total_Amount'].sum()
monthly_sales.plot(kind='line')
plt.title('Monthly Sales Trend')
plt.ylabel('Total Revenue')
plt.xlabel('Month')
plt.show()
```

```python
#Customer Purchase Behavior


# Lifetime value per customer
ltv = df.groupby('Customer_ID')['Total_Amount'].sum().sort_values(ascending=False).head(10)
ltv.plot(kind='bar')
plt.title('Top 10 Customers by Lifetime Value')
plt.ylabel('Total Revenue')
plt.show()

```

```python
#Product Ratings and Review Sentiment

# Review sentiment (assuming sentiment column is available)
sentiment_count = df['Feedback'].value_counts()
sentiment_count.plot(kind='bar')
plt.title('Customer Review Sentiment Distribution')
plt.ylabel('Count of Reviews')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Violin plot for Ratings distribution
sns.violinplot(x=df['Ratings'].dropna())
plt.title('Product Ratings Distribution')
plt.xlabel('Ratings')
plt.show()
```

```python
# Shipping Method and Delivery Analysis

# Shipping method popularity
sns.countplot(x='shipping method', data=df)
plt.title('Shipping Method Popularity')
plt.ylabel('Count of Transactions')
plt.show()

# Delivery time vs Ratings
sns.scatterplot(x='delivary time', y='Ratings', data=df)
plt.title('Delivery Time vs Ratings')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Product Ratings')
plt.show()
```

```python
#Payment Method and Transaction Analysis

# Payment method distribution
sns.countplot(x='Payment_Method', data=df)
plt.title('Payment Method Usage')
plt.ylabel('Count of Transactions')
plt.show()

# Average transaction value by payment method
avg_transaction_value = df.groupby('Payment_Method')['Total_Amount'].mean()
avg_transaction_value.plot(kind='bar')
plt.title('Average Transaction Value by Payment Method')
plt.ylabel('Average Transaction Amount')
plt.show()
```

```python
# feedback Distribution (Violin Plot)
sns.violinplot(x='Customer_Segment', y='Ratings', data=df)
plt.title('Distribution of Rating Across Customer Segments')
plt.show()
```

```python

#Pareto Chart (80/20 Rule for Products or Customers)
product_sales = df.groupby('product name')['Total_Amount'].sum().sort_values(ascending=False)
cumulative_sales = product_sales.cumsum() / product_sales.sum() * 100

fig, ax = plt.subplots()
product_sales.plot(kind='bar', ax=ax)
cumulative_sales.plot(kind='line', color='red', ax=ax, secondary_y=True)
ax.set_ylabel('Total Sales')
ax.right_ax.set_ylabel('Cumulative Percentage')
plt.title('Pareto Chart for Product Sales (80/20 Rule)')
plt.show()
```
